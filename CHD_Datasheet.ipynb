**Plot Continues Variables.**

```python
data.columns
```

```python
cate_val=[]
cont_val=[]

for column in data.columns:
  if data[column].nunique() <=10:
    cate_val.append(column)
  else:
      cont_val.append(column)


```

```python
cate_val
```

```python
cont_val
```

```python
data.hist(cont_val,figsize=(17,15))
plt.show()
plt.tight_layout()
```
